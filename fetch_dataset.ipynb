```python
from nilearn.datasets import fetch_neurovault
from urllib.request import urlopen
import json
```

Fetching

```python
MAX_IMAGES = 5
fmri_scans = fetch_neurovault(max_images=MAX_IMAGES)
```

```python
fmri_scans
```

Now fetching from Cognitive Atlas

```python
cog_concepts = urlopen('https://www.cognitiveatlas.org/api/v-alpha/concept')
cog_concepts = json.loads(cog_concepts.read())
```

```python
len(cog_concepts)
```

We just fetched 886 cognitive concepts. It is upto us which ones to map the neurovault images to.
